```python
def smartdiv(func):
    def inner(x,y):
        if x<y:
            x,y=y,x
        return func(x,y)
    return inner
```

```python
@smartdiv
def div(x,y):
    return x/y
```

```python
import time

def smart(func):
    def cal():                # The wrapper function
        start = time.time()   # Record start time
        func()                # Execute the passed function
        end = time.time()     # Record end time
        final = end - start
        print(f"time to run this code is {final}")
    return cal                # Return the wrapper function

@smart
def demo():
    time.sleep(1)             # Simulate work
    i=0
    while i<1000:
        print(i)
        i=i+1

demo()
```
